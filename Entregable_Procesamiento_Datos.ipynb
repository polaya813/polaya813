Parte 2: Limpieza de Datos
1. Manejo de valores faltantes (NaN):
o Identifica cuántos valores faltantes hay por columna.
o Para la columna Cantidad, imputa los valores faltantes con la mediana de la
columna.
o Para la columna Precio_Unitario, si encuentras faltantes, imputa con la
media.

```python
print("\nValores faltantes por columna ANTES de la imputación:")
print(df_ventas.isnull().sum())
```

```python

# Imputar 'Cantidad' con la mediana
mediana_cantidad = df_ventas['Cantidad'].median()
df_ventas['Cantidad'] = df_ventas['Cantidad'].fillna(mediana_cantidad)

print(f"\nValores NaN en 'Cantidad' imputados con la mediana: {mediana_cantidad}")
```

```python

# Imputar 'Precio_Unitario' con la media si es necesario
if df_ventas['Precio_Unitario'].isnull().any():
    media_precio = df_ventas['Precio_Unitario'].mean()
    df_ventas['Precio_Unitario'] = df_ventas['Precio_Unitario'].fillna(media_precio)
    print(f"Valores NaN en 'Precio_Unitario' imputados con la media: {media_precio}")
```

```python
print("\nValores faltantes por columna DESPUÉS de la imputación:")
print(df_ventas.isnull().sum())
```

Manejo de duplicados:
o Identifica y elimina filas duplicadas, si existen.

```python
duplicados_antes = df_ventas.duplicated().sum()
print(f"\nNúmero de filas duplicadas ANTES de eliminar: {duplicados_antes}")
```

```python
df_ventas.drop_duplicates(inplace=True)
```

```python
duplicados_despues = df_ventas.duplicated().sum()
print(f"Número de filas duplicadas DESPUÉS de eliminar: {duplicados_despues}")
```

```python
print("Filas duplicadas eliminadas.")
```

Corrección de tipos de datos:
o Convierte la columna Fecha a tipo datetime.
o Asegúrate de que Cantidad y Precio_Unitario sean numéricos. (Ya
deberían serlo tras la imputación, pero es buena práctica verificarlo o forzarlo si
fuera necesario).

```python
df_ventas['Fecha'] = pd.to_datetime(df_ventas['Fecha'])
df_ventas['Cantidad'] = pd.to_numeric(df_ventas['Cantidad'])
df_ventas['Precio_Unitario'] = pd.to_numeric(df_ventas['Precio_Unitario'])
print("\nTipos de datos después de la conversión:")
df_ventas.info()
```

```python
# Para demostrar la estandarización, cambiaremos 200 registros de 'Categoria' a minúsculas.
# Primero, asegurémonos de que haya al menos 200 registros en el DataFrame.
if len(df_ventas) >= 200:
    # Seleccionar 200 índices aleatorios
    indices_a_modificar = np.random.choice(df_ventas.index, 200, replace=False)

    # Convertir a minúsculas las categorías en esos índices
    df_ventas.loc[indices_a_modificar, 'Categoria'] = df_ventas.loc[indices_a_modificar, 'Categoria'].str.lower()
    print("Se han modificado 200 registros en la columna 'Categoria' a minúsculas.")
else:
    print(f"El DataFrame tiene menos de 200 registros ({len(df_ventas)}). No se han modificado categorías a minúsculas.")
```

Estandarización de texto:
o La columna Categoria tiene "Electronica" y "electronica". Estandariza a un

```python
print("\nValores únicos en 'Categoria' ANTES de estandarizar:")
print(df_ventas['Categoria'].unique())
```

```python
df_ventas['Categoria'] = df_ventas['Categoria'].str.capitalize()
print("\nValores únicos en 'Categoria' DESPUÉS de estandarizar:")
```

Parte 3: Transformación de Datos y Feature Engineering
1. Crear una columna Ingreso_Total: Multiplica Cantidad por Precio_Unitario.

```python
df_ventas['Ingreso_Total'] = df_ventas['Cantidad'] * df_ventas['Precio_Unitario']
print("\nDataFrame con la nueva columna 'Ingreso_Total':")
print(df_ventas.head())
```

Extraer componentes de la fecha: Crea columnas para Año, Mes y Dia_Semana a
partir de la columna Fecha.

```python
df_ventas['Año'] = df_ventas['Fecha'].dt.year
df_ventas['Mes'] = df_ventas['Fecha'].dt.month
df_ventas['Dia_Semana'] = df_ventas['Fecha'].dt.day_name()
print("\nDataFrame con columnas de fecha extraídas:")
print(df_ventas[['Fecha', 'Año', 'Mes', 'Dia_Semana']].head())
```

Parte 4: Análisis Exploratorio con Visualizaciones
1. Distribución de Ingreso_Total: Crea un histograma y un boxplot para la
columna Ingreso_Total.

```python
plt.figure(figsize=(12, 5))
plt.subplot(1, 2, 1) # 1 fila, 2 columnas, primer gráfico
sns.histplot(df_ventas['Ingreso_Total'], kde=True)
plt.title('Distribución de Ingreso Total')
plt.xlabel('Ingreso Total')
plt.ylabel('Frecuencia')
plt.subplot(1, 2, 2) # 1 fila, 2 columnas, segundo gráfico
sns.boxplot(x=df_ventas['Ingreso_Total'])
plt.title('Boxplot de Ingreso Total')
plt.xlabel('Ingreso Total')
plt.tight_layout() # Ajusta el layout para evitar solapamientos
plt.show()
```

Ingresos por Categoria: Crea un gráfico de barras mostrando el ingreso total
sumado por categoría.

```python
ingresos_por_categoria = df_ventas.groupby('Categoria')['Ingreso_Total'].sum().sort_values(ascending=False)
plt.figure(figsize=(10, 6))
sns.barplot(x=ingresos_por_categoria.index,
y=ingresos_por_categoria.values, hue=ingresos_por_categoria.index, palette='viridis', legend=False)
plt.title('Ingreso Total por Categoría de Producto')
plt.xlabel('Categoría')
plt.ylabel('Ingreso Total Acumulado')
plt.xticks(rotation=45, ha='right')
plt.tight_layout()
plt.show()
```

Ingresos a lo largo del tiempo (por Mes): Crea un gráfico de líneas mostrando la
tendencia de Ingreso_Total por mes. (Nota: Con pocos datos, la tendencia será
simple, pero el ejercicio es válido).

```python
ingresos_por_mes = df_ventas.groupby('Mes')['Ingreso_Total'].sum()
plt.figure(figsize=(10, 6))
ingresos_por_mes.plot(kind='line', marker='o')
plt.title('Tendencia de Ingresos Totales por Mes')
plt.xlabel('Mes')
plt.ylabel('Ingreso Total Acumulado')
plt.xticks(ingresos_por_mes.index) # Asegura que todos los meses se muestren
plt.grid(True)
plt.tight_layout()
plt.show()
```
