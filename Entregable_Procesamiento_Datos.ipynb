Parte 2: Limpieza de Datos
1. Manejo de valores faltantes (NaN):
o Identifica cuántos valores faltantes hay por columna.
o Para la columna Cantidad, imputa los valores faltantes con la mediana de la
columna.
o Para la columna Precio_Unitario, si encuentras faltantes, imputa con la
media.

```python
print("\nValores faltantes por columna ANTES de la imputación:")
print(df_ventas.isnull().sum())
```

```python

# Imputar 'Cantidad' con la mediana
mediana_cantidad = df_ventas['Cantidad'].median()
df_ventas['Cantidad'] = df_ventas['Cantidad'].fillna(mediana_cantidad)

print(f"\nValores NaN en 'Cantidad' imputados con la mediana: {mediana_cantidad}")
```

```python

# Imputar 'Precio_Unitario' con la media si es necesario
if df_ventas['Precio_Unitario'].isnull().any():
    media_precio = df_ventas['Precio_Unitario'].mean()
    df_ventas['Precio_Unitario'] = df_ventas['Precio_Unitario'].fillna(media_precio)
    print(f"Valores NaN en 'Precio_Unitario' imputados con la media: {media_precio}")
```

```python
print("\nValores faltantes por columna DESPUÉS de la imputación:")
print(df_ventas.isnull().sum())
```

Corrección de tipos de datos:
o Convierte la columna Fecha a tipo datetime.
o Asegúrate de que Cantidad y Precio_Unitario sean numéricos. (Ya
deberían serlo tras la imputación, pero es buena práctica verificarlo o forzarlo si
fuera necesario).

```python
df_ventas['Fecha'] = pd.to_datetime(df_ventas['Fecha'])
df_ventas['Cantidad'] = pd.to_numeric(df_ventas['Cantidad'])
df_ventas['Precio_Unitario'] = pd.to_numeric(df_ventas['Precio_Unitario'])
print("\nTipos de datos después de la conversión:")
df_ventas.info()
```

```python
# Para demostrar la estandarización, cambiaremos 200 registros de 'Categoria' a minúsculas.
# Primero, asegurémonos de que haya al menos 200 registros en el DataFrame.
if len(df_ventas) >= 200:
    # Seleccionar 200 índices aleatorios
    indices_a_modificar = np.random.choice(df_ventas.index, 200, replace=False)

    # Convertir a minúsculas las categorías en esos índices
    df_ventas.loc[indices_a_modificar, 'Categoria'] = df_ventas.loc[indices_a_modificar, 'Categoria'].str.lower()
    print("Se han modificado 200 registros en la columna 'Categoria' a minúsculas.")
else:
    print(f"El DataFrame tiene menos de 200 registros ({len(df_ventas)}). No se han modificado categorías a minúsculas.")
```
