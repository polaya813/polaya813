```python
print("\nValores faltantes por columna ANTES de la imputación:")
print(df_ventas.isnull().sum())
```

```python
if 'Cantidad' in df_ventas.columns:
    mediana_cantidad = df_ventas['Cantidad'].median()
    df_ventas['Cantidad'].fillna(mediana_cantidad, inplace=True)
    print(f"\nValores NaN en 'Cantidad' imputados con la mediana: {mediana_cantidad}")
else:
    print("Error: La columna 'Cantidad' no existe en el DataFrame 'df_ventas'.")
    print(f"Columnas disponibles: {df_ventas.columns.tolist()}")
    print("Además, el análisis previo mostró que no hay valores nulos en el DataFrame actual.")
```
