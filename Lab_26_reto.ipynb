```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
from scipy import stats
!pip install unidecode
import unidecode
```

```python
# Cargar el dataset
url = "https://raw.githubusercontent.com/datasets/population/main/data/population.csv"
df_raw = pd.read_csv(url)
df = df_raw.copy()
df.head()
```

Paso 2: Exploración inicial

```python
print("Dimensiones:", df.shape)
df.info()
df.describe(include='all')
```

Paso 3: Identificar y eliminar duplicados
python

```python
print("Duplicados:", df.duplicated().sum())
df = df.drop_duplicates()
```

Paso 4: Manejo de valores nulos

```python
df.isnull().sum()
# Imputar valores numéricos con la media
df['Age'] = df['Age'].fillna(df['Age'].mean())
# Rellenar texto con “Sin datos”
df['Job Title'] = df['Job Title'].fillna("Sin datos")
```
