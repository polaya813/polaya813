Archivo generado con datos simulados

```python
import pandas as pd
import numpy as np

# Generar datos simulados
np.random.seed(0)
num_records = 1000
data = {
    'id': np.arange(1, num_records + 1),
    'name': [f'Cliente {i}' for i in range(1, num_records + 1)],
    'ciudad': np.random.choice(['Ciudad A', 'Ciudad B', 'Ciudad C'], size=num_records),
    'pais': np.random.choice(['Pais X', 'Pais Y', 'Pais Z'], size=num_records)
}

# Crear un DataFrame
df = pd.DataFrame(data)

# Guardar en un archivo CSV
df.to_csv('clientes_raw.csv', index=False)
print("Archivo CSV generado con éxito.")
```

Extract

```python
def extract_clientes(filepath='clientes_raw.csv'):
    """Extract data from a CSV file."""
    try:
        df_clientes_raw = pd.read_csv(filepath)
        print("Datos extraídos con éxito.")
        return df_clientes_raw
    except Exception as e:
        print(f"Error al extraer datos: {e}")
        return None
```

Transform

```python
def transform_clientes(df_raw):
    """Transform raw data to match the dimension table structure."""
    df_transformed = df_raw.copy()
    df_transformed.rename(columns={'id': 'cliente_id_origen', 'name': 'nombre_cliente'}, inplace=True)

    # Añadir claves subrogadas si es necesario
    df_transformed['cliente_key'] = range(1, len(df_transformed) + 1)

    # Seleccionar columnas necesarias
    df_transformed = df_transformed[['cliente_key', 'cliente_id_origen', 'nombre_cliente', 'ciudad', 'pais']]

    print("Datos transformados con éxito.")
    return df_transformed
```

Load

```python
def load_dimension(df, output_filepath='dim_clientes.csv'):
    """Load data into a CSV file."""
    try:
        df.to_csv(output_filepath, index=False)
        print(f"Datos cargados exitosamente en el archivo {output_filepath}.")
    except Exception as e:
        print(f"Error al cargar datos en el archivo: {e}")
```

Ejecución del Pipeline ETL

```python
def main():
    # Extract
    df_clientes_raw = extract_clientes()
    if df_clientes_raw is not None:
        # Transform
        df_dim_clientes = transform_clientes(df_clientes_raw)
        # Load
        load_dimension(df_dim_clientes)

if __name__ == '__main__':
    main()
```
