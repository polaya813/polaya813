### Manejo de valores faltantes

```python
# Porcemtaje de nulos por columnas antes de la imputación
print((df_life.isnull().sum() / len(df_life)) * 100)
```

```python
cols_to_impute_numeric = ['life_expectancy', 'adult_mortality', 'alcohol', 'gdp', 'schooling', 'bmi', 'income_composition_of_resources']
for col in cols_to_impute_numeric:
  df_life[col] = df_life.groupby('status')[col].transform(lambda x: x.fillna(x.median()))
```

```python
# Para otras columnas con muchos nulos y que no son el focoprincipal, podrías eliminarlas o imputar con una constante si esnecesario
# Por simplicidad, nos enfocamos en las anteriores. Si alguna columna aún tiene nulos después de esto, rellenamos con la mediana global.
for col in cols_to_impute_numeric:
  if df_life[col].isnull().any():
    df_life[col].fillna(df_life[col].median(), inplace=True)
```

```python
# Valores faltantes después de la imputación
df_life[cols_to_impute_numeric + ['status']].isnull().sum()
```

# Análisis Estadístico Descriptivo Univariado

## Medidas estadísticas

1. Medidas de Tendencia Central: Calcula la media, mediana y moda.
2. Medidas de Dispersión: Calcula el rango, varianza y desviación estándar.
3. Medidas de Forma: Calcula la asimetría (skewness) y curtosis.

```python
variables_num = ['life_expectancy', 'adult_mortality', 'gdp','schooling']
```

```python
# Llamar la variable
variables_num
```

```python
# crear un dataframe nuevo y vacio para hacer los calculos de meen, media, moda y otros
stats_descriptive = pd.DataFrame()
stats_descriptive
```

```python
for var in variables_num:
  media = df_life[var].mean()
  mediana = df_life[var].median()
  # La moda puede tener múltiples valores, tomamos el primero
  moda = df_life[var].mode()[0] if not df_life[var].mode().empty else np.nan
  rango = df_life[var].max() - df_life[var].min()
  varianza = df_life[var].var()
  std_dev = df_life[var].std()
  asimetria = df_life[var].skew()
  curtosis_val = df_life[var].kurtosis() #Kurtosis de Fisher (exceso)
  stats_descriptive[var] = [media, mediana, moda, rango, varianza, std_dev, asimetria, curtosis_val]
  stats_descriptive.index = ['Media', 'Mediana', 'Moda', 'Rango', 'Varianza', 'Desv. Estándar', 'Asimetría', 'Curtosis']
```
